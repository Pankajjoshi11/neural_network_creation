```python
import tensorflow as tf
```

```python
from tensorflow.keras.datasets import mnist
(training_data, training_labels),(testing_data, testing_labels)=mnist.load_data()
```

```python
training_data.shape
testing_data.shape
```

```python
training_data=training_data/255.0
testing_data = testing_data /255.0
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(
        128,
        activation='relu'
    ),
    #relu= rectified linear function
    tf.keras.layers.Dense(
        10,
        activation= 'softmax'
    )
    #softmax = gives probability of the each number represented and the highest probability one wil be selected
])
```

```python
model.compile(
    optimizer= 'sgd',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
#sparse_categorical_crossentropy = crossentropy function (categorical represents there is more than 2 option i.e. not binary)
#loss= loss of information
#sgd(stoocastic gradiant descent)= fucntion, loss optimizer, s=stocastic==picks up the point which is the farthest from the mean line
```

```python
model.fit(
    training_data,
    training_labels,
    epochs = 5
)
#epohcs= process where entire forward and backward propogation happens in neural network
# 5 epoche means it will propogate and return 5 times and with each round it will imrpove its accuracy and learn from the previous errors
```

```python
#testing
classifications= model.predict(testing_data)
```

```python
classifications[0]
```

```python
import numpy as np
test_predicted_label = np.argmax(classifications[0])
test_predicted_label
```
